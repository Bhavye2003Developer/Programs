```python
import matplotlib.pyplot as plt
import random
```

```python
def make_pts(l,m,number_pts):
    #v = lx + my
    x = []
    y = []
    for i in range(number_pts):
        x_val = random.randint(-(number_pts),number_pts+1)
        y_val = random.randint(-(number_pts),number_pts+1)
        value = (l * x_val) + (m * y_val)
        y.append(value)
        x.append(x_val)
    return x,y

x,y = make_pts(1,1,5)
print(x)
print(y)
```
